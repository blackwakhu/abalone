```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('abalone.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

cols = ['Length','Diameter','Height','Whole weight','Shucked weight', 'Viscera weight', 'Shell weight', 'rings']

for col in cols:
    df[col] = scale.fit_transform(df[col][:,np.newaxis])
```

```python
sns.pairplot(df)
```

```python
corr_col = df.drop(['Sex'], axis=1)
corr_matrix = df.corr()
sns.heatmap(corr_matrix)
```

```python
from sklearn.preprocessing import LabelEncoder
labeller = LabelEncoder()
df['Sex'] = labeller.fit_transform(df['Sex'])
```

```python
plt.scatter(df['Length'], df['Diameter'], c=df['Sex'], cmap='viridis')
plt.xlabel('length'); plt.ylabel('diameter')
```

```python
sns.displot(df['Length'])
```

```python
from sklearn.model_selection import train_test_split
y = df['Sex']
X = df.drop(['Sex'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.75)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

naive = GaussianNB()
randomforest = RandomForestClassifier()
neighbours = KNeighborsClassifier(n_neighbors=8)
models = [naive, randomforest, neighbours]
```

```python
from sklearn.metrics import accuracy_score

for model in models:
    model.fit(X_train, y_train)
    prediction = model.predict(X_test)
    accuracy = accuracy_score(y_test, prediction)
    print(f'{model}: {accuracy}')
```
